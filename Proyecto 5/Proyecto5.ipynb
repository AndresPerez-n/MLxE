```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None) #Mostrar todas las columnas
```

```python
df_ocupados= pd.read_csv('Ocupados.csv',sep=';')
df_ocupados.head()
```

```python
df_Caracteristicas_generales=pd.read_csv('Caracteristicas generales.csv',sep=';')
df_Caracteristicas_generales.head()
```

```python
df_desocupados= pd.read_csv('Desocupados.csv',sep=';')
df_desocupados.head()
```

```python
df_inactivos=pd.read_csv('Inactivos.csv',sep=';')
df_inactivos.head()
```
